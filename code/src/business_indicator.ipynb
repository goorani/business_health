# Business Health Indicator Analysis

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset/fy24c2 business health dataset.csv')
```

```python
df.head()
```

## Calculating business health indicators

```python
df['%grant_health'] = (df['sb_grant_used'] / df['sb_grant_value']) * 100
df['%value_health'] = (df['business_value'] / df['sb_grant_value']) * 100
df['%profit_health'] = (df['business_profit'] / df['sb_grant_value']) * 100
```

```python
df['grant_health'] = ''
df['value_health'] = ''
df['profit_health'] = '' 
df['member_health'] = ''
df['record_health'] = ''
```

### Defining conditional functions

```python
def grant_value_profit_health(health):
    if health >=75:
        return 'Green'
    elif health >=51:
        return 'Yellow'
    else:
        return 'Red'
```

```python
def member_health(drop):
    if drop ==1:
        return 'Yellow'
    if drop < 1:
        return 'Red'
    else:
        return 'Green'
```

```python
def record_health(record):
    if record == 'Yes':
        return 'Green'
    else:
        return 'Yellow'
```

### Assigning mentoring topics based on teh performance indicator

```python
def mentoring_grant_topic(health):
    if health == 'Yellow' or health == 'Red':
        return 'Grant Invenstment: Group have invested all the grants into the group business'
```

```python
def mentoring_value_topic(health):
    if health == 'Yellow' or health == 'Red':
        return 'Business Group Mentoring: Groups have selected and launched profitable and sustainable businesses'
```

```python

def mentoring_profit_topic(health):
    if health == 'Yellow' or health == 'Red':
        return 'Business Group Mentoring: Business owners are maximizing business profit and productivity'
```

```python
def mentoring_member_topic(health):
    if health == 'Yellow' or health == 'Red':
        return 'Business Group Mentoring: Group working together to run the business with the support of their families'

```

```python
def mentoring_record_topic(health):
    if health == 'Yellow' or health == 'Red':
        return 'Record Keeping: Business owners keep accurate records of cash-in, cash-out, assets, debts, and debtors'

        
```

```python
# Apply the function to grant_health
df['grant_health'] = df['%grant_health'].apply(grant_value_profit_health)
```

```python
# Apply the function to grant_health
df['value_health'] = df['%value_health'].apply(grant_value_profit_health)
```

```python
# Apply the function to grant_health
df['profit_health'] = df['%profit_health'].apply(grant_value_profit_health)
```

```python
# Apply the function to member_health
df['member_health'] = df['number_of_bos_dropped'].apply(member_health)
```

```python
# Apply the function to record_health
df['record_health'] = df['records_uptodate'].apply(record_health)
```

```python
df['mentoring_grant_health'] = df['grant_health'].apply(mentoring_grant_topic)
```

```python
df['mentoring_value_health'] = df['value_health'].apply(mentoring_value_topic)
```

```python
df['mentoring_profit_health'] = df['profit_health'].apply(mentoring_profit_topic)
```

```python
df['mentoring_member_health'] = df['member_health'].apply(mentoring_member_topic)
```

```python
df['mentoring_record_health'] = df['record_health'].apply(mentoring_record_topic)
```
